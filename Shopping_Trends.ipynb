```python
#Now Import the Python Libraries for Data Visualzation
# 1.Matplotplib
# 2.Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Which Category has the higest sell?
df['Category'].value_counts().plot(kind="bar",figsize=(15,5),color=['#ff3399','#338aff','#33f3ff','#ffb233'])
plt.title("Category with highest sell")
plt.xlabel("Category")
plt.ylabel("Values")
plt.xticks(rotation=0)
plt.show()
```

```python
#Which iterms is most sold check?
df['Item Purchased'].value_counts().plot(kind="pie",autopct="%.2f",figsize=(15,10))
plt.title("Items Sold")
plt.xlabel("Items")
plt.ylabel("")
plt.show()
```

```python
#Which products will most sold in each city?
df_grouped = df.groupby(['Location', 'Category']).size().unstack()
df_grouped.plot(kind="bar", figsize=(12, 6), colormap="tab10")
plt.title("Product Sold by City and Category")
plt.xlabel("Cities")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=90)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Which type of Payment method most used?
df['Payment Method'].value_counts().plot(kind="pie",autopct="%.2f",figsize=(15,10))
plt.title("Most used payment method")
plt.xlabel("Payment Method")
plt.ylabel("")
plt.show()
```

```python
#Which item has most ratings?
df_grouped=df.groupby("Item Purchased")['Review Rating'].sum()
df_grouped.plot(kind='bar',figsize=(20,10))
plt.title("Items with thier Ratings")
plt.xlabel("Items")
plt.ylabel("Ratings")
plt.xticks(rotation=45)
plt.show()
```
