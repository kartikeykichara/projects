```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.How many products sold in each city
df_grouped=df.groupby('Location')['Number of products sold'].sum()
df_grouped.plot(kind='bar',color=['#33feff','#ff6333','#ffc933','#ff33ae','#33bfff'])
plt.title('Products Sold In Each City',fontsize=15)
plt.xlabel('Cities',fontsize=15)
plt.ylabel('Number of products sold',fontsize=15)
plt.xticks(rotation=0)
```

```python
# 2.How many products Sold#
df['Product type'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title('Products Sold',fontsize=15)
plt.xlabel('Product Type',fontsize=15)
plt.xticks(rotation=0)
```

```python
df_grouped=df.groupby('Stock levels')['Order quantities'].sum()
df_grouped.plot(kind='hist')
plt.title('Stock level effect order quantities')
plt.xlabel('Stock level')
plt.ylabel('Order quantities')
plt.xticks(rotation=0)
```
