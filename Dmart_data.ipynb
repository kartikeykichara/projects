```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="most_frequent")
df['Brand']=si.fit_transform(df[['Brand']]).ravel()
df['Name']=si.fit_transform(df[['Name']]).ravel()
df['Category']=si.fit_transform(df[['Category']]).ravel()
df['SubCategory']=si.fit_transform(df[['SubCategory']]).ravel()
df['Quantity']=si.fit_transform(df[['Quantity']]).ravel()
df['Description']=si.fit_transform(df[['Description']]).ravel()
df['BreadCrumbs']=si.fit_transform(df[['BreadCrumbs']]).ravel()
```

```python
df.isnull().sum()
```

```python
si=SimpleImputer(strategy="mean")
df['Price']=si.fit_transform(df[['Price']]).ravel()
df['DiscountedPrice']=si.fit_transform(df[['DiscountedPrice']]).ravel()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_grouped=df.groupby('Category')['DiscountedPrice'].sum()
df_grouped.plot(kind="bar",figsize=(16,9))
plt.title("Brand has highest Discount")
plt.xlabel("Category",fontsize=20)
plt.ylabel("Discount",fontsize=17)
plt.xticks(rotation=45)
plt.show()
```
