```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_grouped=df.groupby('Product Name')['Rating'].sum()
df_grouped.plot(kind="bar",color=['#ff3387','#ff6e33','#33ffbb','#e333ff','#ff3358'])
plt.title("Product has highest rating ")
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Brand'].value_counts().plot(kind="pie",autopct="%.2f",figsize=(18,9))
plt.title("Ratings According To Brand")
plt.xlabel("Brand")
# plt.ylabel("Rating")
# plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.show()
```

```python
df_grouped=df.groupby('Product Name')['Price'].sum()
df_grouped.plot(kind="bar",figsize=(10,5),color=['#33ffe3','#3386ff','#ff338d','#33ffb5','#f033ff'])
plt.title("Highest price")
plt.xlabel("Product Name")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
df_grouped=df.groupby('Brand')['Product Name'].count()
df_grouped.plot(kind="barh",color=['#33c1ff','#ff3383','#33ff80','#3399ff','#ffa833'])
plt.title("Brand has Highest product sold")
plt.xlabel("")
plt.ylabel("Values")
plt.show()
```
