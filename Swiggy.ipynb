```python
#Now Import the Python Libraries for Data Visualzation
# 1.Matplotplib
# 2.Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Which city has highest food rating
Dataframe_groupedby=Dataframe.groupby('City')['Total ratings'].count()
Dataframe_groupedby.plot(kind="bar",figsize=(12,5))
plt.title('City has Highest Food Rating',fontsize=15)
plt.xlabel('City',fontsize=15)
plt.ylabel('Total Ratings',fontsize=15)
plt.show()
```

```python
#2.To Find the top 10 most ordered restaurants
restaurant_col = 'Restaurant'  
if restaurant_col not in Dataframe.columns:
    print("Error: Restaurant column not found! Please check column names.")
else:
    restaurant_counts = Dataframe[restaurant_col].value_counts()
    
    top_10 = restaurant_counts.head(10)

    plt.figure(figsize=(14, 5))

    # Plot for most ordered restaurants
    plt.subplot(1, 2, 1)
    sns.barplot(x=top_10.values, y=top_10.index, palette="Blues_r")
    plt.title("Top 10 Most Ordered Restaurants")
    plt.xlabel("Number of Orders")
    plt.ylabel("Restaurant Name")
    
    plt.show()
```

```python
#3.Which Food Type is most ordered
food_type_col='Food type'
if food_type_col not in Dataframe.columns:
    print("Error: food type  not found! Please check column names.")
else:
    food_type_counts=Dataframe[food_type_col].value_counts()

    top_5_food_type=food_type_counts.head(5)

    plt.figure(figsize=(10,5))

    top_5_food_type.plot(kind="bar",color=['#33ffec','#3393ff','#ca33ff','#ff3386','#9fff33'])
    plt.title("Top 5 Most Ordered Food Type")
    plt.xlabel("Food type")
    plt.ylabel("No of Orders")

    plt.xticks(rotation=0)
    
    plt.show()
```

```python
#4.Top 10 Restaurant by Ratings
ratings_col = 'Avg ratings'
restaurant_col = 'Restaurant'

if ratings_col not in Dataframe.columns or restaurant_col not in Dataframe.columns:
    print(" Avg ratings or Restaurant' column not found! Please check column names.")
else:
    top_restaurants = Dataframe[[restaurant_col, ratings_col]].drop_duplicates().sort_values(by=ratings_col, ascending=False).head(10)

    plt.figure(figsize=(10, 5))
    top_restaurants.plot(kind="barh", x=restaurant_col, y=ratings_col, color="pink", legend=False)

    plt.xlabel("Average Rating")
    plt.ylabel("Restaurant Name")
    plt.title("Top 10 Restaurants by Ratings")

plt.show()
```
