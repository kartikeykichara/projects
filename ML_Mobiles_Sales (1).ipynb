```python
#Convert Categorical Data into Numerical Data  By Encoding
```

```python
Data_Frame.head(5)
```

```python
#import Label Encoder with help of Scikit Learn Library
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
Data_Frame['Product']=lb.fit_transform(Data_Frame['Product'])
Data_Frame['Brand']=lb.fit_transform(Data_Frame['Brand'])
Data_Frame['Product Code']=lb.fit_transform(Data_Frame['Product Code'])
Data_Frame['Product Specification']=lb.fit_transform(Data_Frame['Product Specification'])
Data_Frame['Customer Name']=lb.fit_transform(Data_Frame['Customer Name'])
Data_Frame['Customer Location']=lb.fit_transform(Data_Frame['Customer Location'])
Data_Frame['Region']=lb.fit_transform(Data_Frame['Region'])
Data_Frame['Processor Specification']=lb.fit_transform(Data_Frame['Processor Specification'])
Data_Frame['RAM']=lb.fit_transform(Data_Frame['RAM'])
Data_Frame['ROM']=lb.fit_transform(Data_Frame['ROM'])
Data_Frame['Inward Date']=lb.fit_transform(Data_Frame['Inward Date'])
Data_Frame['Dispatch Date']=lb.fit_transform(Data_Frame['Dispatch Date'])
```

```python
Data_Frame.head(5)
```

```python
#Data Dividation
from sklearn.model_selection import train_test_split
```

```python
#Input Data target Data
x=Data_Frame.drop(columns=['Region'])
y=Data_Frame['Region']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Now import the Machine Learning Model
from sklearn.ensemble import RandomForestClassifier
```

```python
#Create the object of Model
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Print the accuracy of our Dataset using Random Forest Classifier Model 
Prediction=accuracy_score(y_test,y_pred)
print(Prediction)
```
